```python
frame.columns.duplicated()
```

```python
frame = frame.iloc[:, ~frame.columns.duplicated()].drop(columns=['label'])
```

```python
frame.info(verbose=True)
```

```python
frame.to_csv(path.join('..', 'processed', '1-covid','mathematical-representation.csv'), index=False)
```

```python
# Feature selection with ANOVA

from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import f_classif
```

```python
frame.iloc[:5, 1:152]
```

```python
selector = SelectKBest(score_func=f_classif, k=20)

fit = selector.fit(frame.iloc[:, 1:152],frame.iloc[:, 152])
```

```python
new_data = fit.transform(frame.iloc[:, 1: 152])

new_data.shape
```

```python
selected_columns = fit.get_support()

len(selected_columns)
```

```python
df_anova = frame.iloc[:, 1: 152].iloc[:,selected_columns]
df_anova.head()
```

```python
anova_columns = list()

for col in df_anova.columns:
    anova_columns.append(col)

anova_columns.insert(0, 'nameseq')
anova_columns.append('class')

print(len(anova_columns), anova_columns)
```

```python
df_anova['nameseq'] = frame['nameseq']
df_anova['class']   = frame['class']

# reposicionar as colunas somente por conviniência
df_anova = df_anova[anova_columns]

df_anova.head()
```
